```python
#%%
import cv2
import time
from keras.models import load_model
import numpy as np
model = load_model('keras_model.h5', compile = 'False')
cap = cv2.VideoCapture(0)
data = np.ndarray(shape=(1,224,224,3), dtype=np.float32)
import random
possible_input = ['rock','paper','scissor']
computer_input = random.choice(possible_input)
game = 1
user = 0
computer = 0
started = False

while game <= 5:
    ret, frame = cap.read()
    resized_frame = cv2.resize( frame, (224, 224), interpolation = cv2.INTER_AREA)
    image_np = np.array(resized_frame)
    normalized_image = (image_np.astype(np.float32) / 127.0) - 1 # Normalize the image
    data[0] = normalized_image
    prediction = model.predict(data)
    cv2.imshow( 'frame', frame)
    #image = cv2.putText( frame, 'Total game: ', (5 ,5) ,'FONT_HERSHEY_SIMPLEX',1,(0,0,255), 1, cv2.LINE_AA,0)
    #cv2.imshow(frame, image)
    #image = cv2.putText(image, text, org, font, fontScale, color, thickness, cv2.LINE_AA, False)
    # Press q to close the window
    
    if started == False:
        if(cv2.waitKey(1) & 0xFF == ord('p')) :
            started = True
    if started == True:

        if (cv2.waitKey(1) & 0xFF == ord('q')) :
            break
        elif prediction[0][0] > 0.7:
            user_input = 'rock'
            print(f"\nYou chose {user_input}, computer chose {computer_input}.\n")
            if user_input == computer_input:
                print("Game is Tie") 
            elif user_input == 'rock' and computer_input == 'scissor':
                print("YOU WON!!COMPUTER LOSE, ROCK SMASHES SCISSOR")
                user = user + 1
            elif user_input == 'rock' and computer_input == 'paper':
                print("COMPUTER WON!!YOU LOSE, PAPER COVERS ROCK")
                computer = computer + 1   
            game = game + 1 
        elif prediction[0][1] > 0.7:
            user_input = 'paper'
            print(f"\nYou chose {user_input}, computer chose {computer_input}.\n") 
            if user_input == computer_input:
                print("Game is Tie")  
            elif user_input == 'paper' and computer_input == 'rock':
                print("YOU WON!!COMPUTER LOSE,PAPER COVERS ROCK")
                user = user + 1
            elif user_input == 'paper' and computer_input == 'sciccor':
                print("COMPUTER WON!!YOU LOSE, SCISSOR CUTS THE PAPER") 
                computer = computer + 1   
            game = game + 1
        elif prediction[0][2] > 0.7:
            user_input = 'scissor'
            print(f"\nYou chose {user_input}, computer chose {computer_input}.\n")
            if user_input == computer_input:
                print("Game is Tie")   
            elif user_input == 'scissor'and computer_input == 'paper':
                print("YOU WINS!!COMPUTER LOSE, SCISSOR CUTS THE PAPER")
                user = user + 1
            else:
                print("COMPUTER WINS!!YOU LOSE, ROCK SMASHES SCISSOR")   
                computer = computer + 1   
            game = game + 1
        started = False

if user > computer: print('The winner is User')
else: print('The winner is computer')
#cv2.putText(frame, str('Total game: ', game), (1 , 224), 'FONT_HERSHEY_SIMPLEX,FONT_HERSHEY_PLAIN',10, 1, (0,0,255), 1, cv2.LINE_AA)
print ('Total game: ', game) 
print ('User score: ', user)
print ('Computer score: ', computer)  
print( 'Game tie: ',(game - user - computer))        
      
                
time.sleep(3)    
cap.release()
cv2.destroyAllWindows()

```
