```python
import requests
badges = requests.get('https://api.stackexchange.com/2.3/badges?page=1&fromdate=1650240000&todate=1650326400&order=asc&sort=name&site=stackoverflow')
print(badges)
```

```python
Firstbadges = badges.text
from bs4 import BeautifulSoup
soup = BeautifulSoup(Firstbadges,'html.parser')
print(soup)
```
